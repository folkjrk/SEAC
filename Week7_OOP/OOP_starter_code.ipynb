3. Building an object <br>
The perfect object consists of its properties/attributes, behavior (methods), and identity. So far, we have got to build an object Cust_A with identity and attributes. We will be adding some methods to this object.

4. Inheritance. <br>
Let's creating the class to update the loan products information. We will need some reference key the customer database. <br> The customer database will be the parent class. Loan information is the child class, invoking some of customer attributes.

````python
# parent class
class Customer_info(object):
  # the constructor
  def __init__(self, name, age, occupation):
    self.name = name
    self.age = age
    self.occupation = occupation

  # parent class methods
  def display_info(self):
    print(f'Customer name: {self.name}')
    print(f'Customer age: {self.age}')
    print(f'Customer job: {self.occupation}')

# child class
class Loan_info(Customer_info):
  def __init__(self, name, age, occupation, loan_cat, out_amt, interest, terms):
    self.loan_cat = loan_cat
    self.out_amt = out_amt
    self.interest = interest 
    self.terms = terms 
    # invoking customer info
    Customer_info.__init__(self, name, age, occupation)

  # child class methods
  def display_loan_info(self):
    print(f'Customer name: {self.name}')
    print(f'Loan type: {self.loan_cat}')

  def get_installment(self):
    self.totalamt = (self.out_amt*(1+(self.terms*(self.interest/12))))
    self.installment = self.totalamt/self.terms
    print(f'Installment per month: {self.installment}')

a = Loan_info('A', 40, 'Employee', 'Mortage', 1000000, 0.035, 120)

a.display_info()
print()
a.get_installment()
print()
a.display_loan_info()
````

5. Polymorphism <br>
Let's say we have multiple types of deposit products and thier interest rates are different. Even though all products can be linked to the same object, we could build a specific method for a particular product.

````python
class deposit:
  def __init__(self, deposit_amt):
    self.deposit_amt = deposit_amt

class savings(deposit):
    def __init__(self, deposit_amt):
      self.deposit_amt = deposit_amt
      deposit.__init__(self, deposit_amt)
    def get_interest(self):
      interest = 0.025
      yr3_amount = self.deposit_amt * ((1+interest) ** 3)
      print(f'Amount after three years: {yr3_amount}')
````

````python
d = savings(100)
d.get_interest()
````

6. Encapsulation and Abstraction <br>
An OOP object encapsulates attributes and methods together. We could make some methods to be invisible to public users. Hence the method is protected. 

````python
# Python program to 
# demonstrate private methods
  
# Creating a Base class 
class Base: 
  
    # Declaring public method
    def fun(self):
        print("Public method")
  
    # Declaring private method
    def __fun(self):
        print("Private method")
  
# Creating a derived class 
class Derived(Base): 
    def __init__(self):  
        # Calling constructor of 
        # Base class 
        Base.__init__(self) 
          
    def call_public(self): 
        # Calling public method of base class
        print("\nInside derived class")
        self.fun()
          
    def call_private(self):
        # Calling private method of base class
        self.__fun()
  
# Driver code 
obj1 = Base()
  
# Calling public method
obj1.fun()
  
obj2 = Derived()
obj2.call_public()
  
# Uncommenting obj1.__fun() will 
# raise an AttributeError 
  
# Uncommenting obj2.call_private() 
# will also raise an AttributeError
````
