Imagine we are going to build the bank loan tracking systems. Let's say we have two clients, A and B, which have different loan plans, interest rates, and payback terms. We will try to do it using typical Python scripting first

```python
# define client A
client_a = {'Name'        : 'A', 
            'Occupation'  : 'Employee',
            'Age'         : 40,
            'Mortgage'    : {'Having': True, 
                             'Rate': 3.5, 
                             'OutstandingAmount': 2100000, 
                             'Terms': 300},
            'CreditCard'  : {'Having': False},
            'PersonalLoan': {'Having': True, 
                             'Rate': 6.5, 
                             'OutstandingAmount': 210000, 
                             'Terms': 120}}

# define client B
client_b = {'Name'        : 'B', 
            'Occupation'  : 'Retired',
            'Age'         : 65,
            'Mortgage'    : {'Having': False}, 
            'CreditCard'  : {'Having': True,
                             'OutstandingAmount': 7920},
            'PersonalLoan': {'Having': True, 
                             'Rate': 5.5, 
                             'OutstandingAmount': 210000, 
                             'Terms': 100}}

# combine A and B into the same database
customer_data = {'A': client_a,
                 'B': client_b}
```

What do we do if we want to update the Mortage load status of the client A
- Terms to be 299
- Outstanding amount to be 2090000

```python
customer_data['A']['Mortgage']['Terms'] = 299
customer_data['A']['Mortgage']['OutstandingAmount'] = 2090000

print(customer_data)
```
