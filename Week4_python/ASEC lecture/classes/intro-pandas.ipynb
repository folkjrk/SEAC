```python
import pandas as pd
```

```python
# read dataset
df = pd.read_csv('https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv')
```

```python
for c in ['region-code', 'sub-region-code', 'intermediate-region-code']:
    df[c] = df[c].fillna(999.0).astype(int).astype(str)
```
